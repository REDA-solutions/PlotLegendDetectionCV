```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
clean_image = cv2.imread("../raw_data/helvetios_challenge_dataset_training/images/20220915194538484039.png")
dirty_image = cv2.imread("../raw_data/helvetios_challenge_dataset_training/images/20220915194541155437.png")
plt.figure(figsize=(20,6))
plt.subplot(121)
plt.imshow(clean_image)
plt.subplot(122)
plt.imshow(dirty_image)
plt.show()
```

## Grayscale Histogram

```python
clean_image_g = cv2.cvtColor(clean_image, cv2.COLOR_BGR2GRAY)
dirty_image_g = cv2.cvtColor(dirty_image, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(12,3))
ax = plt.subplot(121)
histr_clean = cv2.calcHist([clean_image_g],[0],None,[256],[0,256])
plt.plot(histr_clean)
plt.subplot(122, sharey=ax)
histr_dirty = cv2.calcHist([dirty_image_g],[0],None,[256],[0,256])
plt.plot(histr_dirty)
plt.show()
```

```python
plt.figure(figsize=(12,3))
ax = plt.subplot(121)
plt.ylim(-50, 10000)
histr_clean = cv2.calcHist([clean_image_g],[0],None,[256],[0,256])
plt.plot(histr_clean)
plt.subplot(122, sharey=ax)
histr_dirty = cv2.calcHist([dirty_image_g],[0],None,[256],[0,256])
plt.plot(histr_dirty)
plt.show()
```
