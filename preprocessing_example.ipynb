# Example: How to use the Preprocessor

```python
from preprocessor import Preprocessor
import cv2
import matplotlib.pyplot as plt
import os
```

```python
path = r'raw_data\helvetios_challenge_dataset_training\images'
img_names = ['20220915194538484039.png', '20220915194540522606.png', '20220915194541155437.png', '20220915194542124279.png']
imgs = [cv2.imread(path + '\\' + img_name) for img_name in img_names]
```

```python
p = Preprocessor(grayscale=True, noise=(True, 5, "gaussian"), adapt_threshold=(True, 25, 2))
```

```python
plt.figure(figsize=(20,25))

for i in range(len(imgs)):
    img = imgs[i]
    plt.subplot(4, 2, i*2+1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB), cmap='Greys_r')
    img = p.preprocess(img)
    plt.subplot(4, 2, i*2+2)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB), cmap='Greys_r')
```

```python
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            images.append(img)
    return images
```
